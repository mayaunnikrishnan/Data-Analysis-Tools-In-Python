# **Matplotlib**
Matplotlib is a popular Python library for creating 2D plots and graphs. The matplotlib.pyplot module is a part of Matplotlib and provides a high-level interface for creating a wide range of plots, from simple line plots to complex visualizations

### **Basic Line Plot:**

Let's start with a simple line plot.

Line plots are commonly used in various fields and situations to represent data over time or to show the relationship between two variables.

 You can create a basic line plot by specifying x and y data points and then use plt.plot() to plot them:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 12, 6]

# Create a line plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
```

```python
#Customizing the Plot:
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 8, 12, 6]
y2 = [5, 8, 12, 10, 14]

# Create line plots with labels and customizations
plt.plot(x, y1, label='Line 1', linestyle='--', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', linestyle='-.', color='green', marker='s')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')
plt.legend()

# Show the plot
plt.grid(True)  # Add gridlines
plt.show()
```

##SUBPLOTS

subplots method provides a way to plot multiple plots on a single figure. Given the number of rows and columns , it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax .

```python
# SUBPLOTS
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
y5 = -np.log(x + 0.1)

# Create subplots
plt.figure(figsize=(15, 5))

# Plot on the first subplot
plt.subplot(1, 5, 1)
plt.plot(x, y1, label='sin(x)', color='blue')
plt.title('Sine Function')
plt.legend()

# Plot on the second subplot
plt.subplot(1, 5, 2)
plt.plot(x, y2, label='cos(x)', color='red')
plt.title('Cosine Function')
plt.legend()

# Plot on the third subplot
plt.subplot(1, 5, 3)
plt.plot(x, y3, label='tan(x)', color='green')
plt.title('Tangent Function')
plt.legend()

# Plot on the fourth subplot
plt.subplot(1, 5, 4)
plt.plot(x, y4, label='exp(x)', color='purple')
plt.title('Exponential Function')
plt.legend()

# Plot on the fifth subplot
plt.subplot(1, 5, 5)
plt.plot(x, y5, label='-log(x + 0.1)', color='orange')
plt.title('Negative Logarithm')
plt.legend()

# Add labels to the x-axis and y-axis of the entire figure
plt.suptitle('Subplots with Line Plots')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
y5 = -np.log(x + 0.1)

# Create subplots in a 2x2 grid
plt.figure(figsize=(10, 8))

# Plot on the first subplot (top-left)
plt.subplot(3, 2, 1)
plt.plot(x, y1, label='sin(x)', color='blue')
plt.title('Sine Function')
plt.legend()

# Plot on the second subplot (top-right)
plt.subplot(3, 2, 2)
plt.plot(x, y2, label='cos(x)', color='red')
plt.title('Cosine Function')
plt.legend()

# Plot on the third subplot (bottom-left)
plt.subplot(3, 2, 3)
plt.plot(x, y3, label='tan(x)', color='green')
plt.title('Tangent Function')
plt.legend()

# Plot on the fourth subplot (bottom-right)
plt.subplot(3, 2, 4)
plt.plot(x, y4, label='exp(x)', color='purple')
plt.title('Exponential Function')
plt.legend()

plt.subplot(3, 2, 5)
plt.plot(x, y5, label='-log(x + 0.1)', color='orange')
plt.title('Negative Logarithm')
plt.legend()


# Add labels to the x-axis and y-axis of the entire figure
plt.suptitle('Subplots with Line Plots (2x2 Grid)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

## **Bar plot**
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 50]

# Create a bar plot
plt.bar(categories, values, color='skyblue')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Show the plot
plt.show()
```

##Stacked barplots
Stacked barplots are a type of barplot in which multiple bars are stacked on top of each other to represent the total value.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [25, 40, 30, 50]
values2 = [15, 20, 35, 45]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the first set of data
ax.bar(categories, values1, label='Dataset 1', color='lightblue')

# Plot the second set of data on top of the first
ax.bar(categories, values2, label='Dataset 2', bottom=values1, color='lightgreen')

# Add labels and a legend
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot Example')
plt.legend()

# Show the plot
plt.show()
```

## **histogram:**
A histogram is a graphical representation of the distribution of numerical data. it shows the frequency of occurrence of each value in a dataset.

## **Pie chart**
A pie chart is a type of data visualization that is used to represent categorical data in a circular graph. It is called a "pie" chart because it resembles a pie that has been sliced into segments, with each segment representing a different category or value. Pie charts are useful for showing the relative proportions or percentages of different categories within a whole.

## **Scatter plot**
A scatter plot is a type of data visualization that is used to display individual data points in a two-dimensional (2D) Cartesian coordinate system. It is particularly useful for visualizing the relationship between two continuous variables. Each data point in a scatter plot is represented by a point on the graph, and the position of each point is determined by the values of the two variables being compared.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the scatter plot
x = np.random.rand(50)  # Random x values
y = 2 * x + 1 + np.random.rand(50)  # Linear relationship with noise

# Create a scatter plot
plt.scatter(x, y, marker='o', color='blue', label='Data Points')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show the plot
plt.legend()
plt.grid(True)
plt.show()
```

## **Contour Plot**
A contour plot, also known as a contour map or contour chart, is a type of data visualization used to display three-dimensional data in two dimensions. It represents the variation in a continuous variable across a two-dimensional surface by showing contour lines or filled regions that connect points with the same value. Contour plots are commonly used in fields such as geography, geology, meteorology, and engineering to visualize data that varies over a surface.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for the contour plot
x = np.linspace(-2, 2, 100)  # X coordinates
y = np.linspace(-2, 2, 100)  # Y coordinates
X, Y = np.meshgrid(x, y)     # Create a grid of coordinates
Z = np.sin(np.sqrt(X**2 + Y**2)) / (np.sqrt(X**2 + Y**2) + 1e-6)  # Example function

# Create a colorful contour plot
contour = plt.contourf(X, Y, Z, levels=20, cmap='viridis')  # Using 'viridis' colormap

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Colorful Contour Plot Example')

# Add a color bar for reference
cbar = plt.colorbar(contour)
cbar.set_label('Z-values')

# Show the plot
plt.grid(True)
plt.show()
```

## **Quiver Plot**

A quiver plot is a type of data visualization that is used to represent vector fields. It is particularly useful for visualizing the direction and magnitude of vectors at different points in a two-dimensional or three-dimensional space. Quiver plots are commonly used in physics, engineering, fluid dynamics, and other fields to visualize phenomena involving vector quantities.








```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for the quiver plot
x = np.linspace(-2, 2, 10)  # X coordinates
y = np.linspace(-2, 2, 10)  # Y coordinates
X, Y = np.meshgrid(x, y)    # Create a grid of coordinates

# Define vector components (direction and magnitude)
U = np.cos(X)  # X-component of vectors
V = np.sin(Y)  # Y-component of vectors

# Create a quiver plot
plt.quiver(X, Y, U, V, scale=20, color='blue', width=0.005)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quiver Plot Example')

# Show the plot
plt.grid(True)
plt.show()
```

## **Box Plot**

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key statistical properties, including the median, quartiles, and potential outliers. Box plots are particularly useful for visualizing the spread and skewness of data, making it easier to identify central tendencies and variations.

```python
import matplotlib.pyplot as plt
import numpy as np
data=[np.random.normal(0,1,100),np.random.normal(2,1,100),np.random.normal(-2,1,100)]
plt.boxplot(data,vert=False,labels=["Dataset1","Dataset2","Dataset3"])
plt.xlabel("values")
plt.ylabel("Datasets")
plt.title("Box plot example")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=[np.random.normal(0,1,100),np.random.normal(2,1,100),
      np.random.normal(-2,1,100)]
box_colors=["lightblue","lightgreen","lightcoral"]
boxplot=plt.boxplot(data,vert=False,labels=["Dataset1","Dataset2","Dataset3"],patch_artist=True)
for box,color in zip(boxplot['boxes'],box_colors):
    box.set(facecolor=color)
whisker_colors=["blue","green","red"]
median_colors=["darkblue",'darkgreen','darkred']
outlier_colors=["blue",'green','red']
for element,color in zip(["whiskers",'medians','fliers'],[whisker_colors,median_colors,outlier_colors]):
    for item,col in zip(boxplot[element],color):
        plt.setp(item,color=col)
plt.xlabel("values")
plt.ylabel("Datasets")
plt.title("Colored Box plot example")
plt.grid(True)
plt.show()
```

## **Violin Plot**

A violin plot is a data visualization that combines aspects of a box plot and a kernel density plot. It is used to visualize the distribution of a dataset and provides insights into its probability density at different values. Violin plots are particularly useful for comparing the distribution of multiple datasets or groups.

note:

The function np.random.seed() is used to set the random seed for the NumPy random number generator. The random seed is an integer value that is used as the starting point for generating random numbers. By setting the seed, you can ensure that the sequence of random numbers generated by the NumPy random functions will be the same every time you run the code, making the results reproducible.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data_to_plot)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = [np.random.normal(0, 1, 100),  # Dataset 1
        np.random.normal(2, 1, 100),  # Dataset 2
        np.random.normal(-2, 1, 100)]  # Dataset 3

# Define custom colors for the violins
violin_colors = ['lightblue', 'lightgreen', 'lightcoral']

# Create a colored violin plot
violin = plt.violinplot(data, showmedians=True, showextrema=False)

# Customize the violins' colors
for i, pc in enumerate(violin['bodies']):
    pc.set_facecolor(violin_colors[i])

# Add labels and a title
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Colored Violin Plot Example')

# Show the plot
plt.grid(True)
plt.show()
```

## **3D Contour Plot**

A 3D contour plot is a visualization technique that displays the contour lines of a three-dimensional surface in a two-dimensional plane. It is useful for visualizing the relationships between three continuous variables and identifying regions of constant or changing values in a 3D dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of X and Y values
X, Y = np.meshgrid(x, y)

# Define a function to calculate Z values (example function)
Z = np.sin(np.sqrt(X**2 + Y**2)) / (np.sqrt(X**2 + Y**2) + 1e-6)

# Create a 3D contour plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
contour = ax.contour(X, Y, Z, cmap='viridis')

# Add labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Contour Plot Example')

# Add a color bar for reference
cbar = fig.colorbar(contour)
cbar.set_label('Z-values')

# Show the plot
plt.show()
```

## **3D Wireframe plot**

A 3D wireframe plot is a visual representation of a three-dimensional surface using lines to connect the data points, creating a wireframe-like structure. It's useful for displaying the shape and structure of a 3D surface, allowing you to visualize how the data changes in three dimensions.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of X and Y values
X, Y = np.meshgrid(x, y)

# Define a function to calculate Z values (example function)
Z = np.sin(np.sqrt(X**2 + Y**2)) / (np.sqrt(X**2 + Y**2) + 1e-6)

# Create a 3D wireframe plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, cmap='viridis')

# Add labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Wireframe Plot Example')

# Show the plot
plt.show()
```

## **3D Surface plot**

A 3D surface plot is a visualization technique used to represent a three-dimensional surface using colors to indicate variations in data values. It's an effective way to visualize complex relationships between three continuous variables and gain insights into the shape and structure of the surface.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of X and Y values
X, Y = np.meshgrid(x, y)

# Define a function to calculate Z values (example function)
Z = np.sin(np.sqrt(X**2 + Y**2)) / (np.sqrt(X**2 + Y**2) + 1e-6)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Surface Plot Example')

# Add a color bar for reference
cbar = fig.colorbar(surface)
cbar.set_label('Z-values')
#to save the plot
plt.savefig('my_plot.png')
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  # Import Axes3D for 3D plotting

# Define the range of values for the X and Y axes
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Define the function (example: a sine wave)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels to the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Customize colormap and colorbar
cbar = fig.colorbar(surface)
cbar.set_label('Z-values')

# Set a title
plt.title('3D Surface Plot Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random sample data
np.random.seed(42)
data = [np.random.normal(0, 1, 100),
        np.random.normal(2, 1, 100),
        np.random.normal(4, 1, 100)]

# Create a box plot
boxprops = dict(linewidth=2, color='blue')
whiskerprops = dict(linewidth=2, linestyle='--', color='green')
flierprops = dict(marker='o', markersize=8, markerfacecolor='red', markeredgecolor='red')

plt.boxplot(data, labels=['Dataset 1', 'Dataset 2', 'Dataset 3'], boxprops=boxprops, whiskerprops=whiskerprops, flierprops=flierprops)

# Customize labels and title
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data for the contour plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create custom contour levels
contour_levels = np.linspace(-1, 1, 11)

# Create a colormap for the contour plot
cmap = plt.cm.get_cmap('viridis', len(contour_levels))

# Create the contour plot
plt.figure(figsize=(8, 6))  # Set the figure size
contour = plt.contourf(X, Y, Z, levels=contour_levels, cmap=cmap)
plt.colorbar(contour, label='Z-Values', ticks=contour_levels)

# Customize contour labels using manual placement and formatting
for level in contour_levels:
    plt.text(0.1, 0.1, f'{level:.1f}', transform=plt.gca().transAxes, color='white', fontsize=8)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Contour Plot')

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data for the plots
x = np.linspace(0, 10, 100)
y_line = np.sin(x)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
categories = ['Category A', 'Category B', 'Category C']
values = [3, 7, 2]

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot a line plot in the top-left subplot
axes[0, 0].plot(x, y_line, color='blue')
axes[0, 0].set_title('Line Plot')

# Plot a scatter plot in the top-right subplot
axes[0, 1].scatter(x_scatter, y_scatter, color='green', marker='o', s=30)
axes[0, 1].set_title('Scatter Plot')

# Plot a bar plot in the bottom-left subplot
axes[1, 0].bar(categories, values, color='orange')
axes[1, 0].set_title('Bar Plot')

# Leave the bottom-right subplot empty

# Adjust subplot spacing
plt.tight_layout()

# Show the plot
plt.show()
```
