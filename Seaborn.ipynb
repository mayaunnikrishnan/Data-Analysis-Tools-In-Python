# **Seaborn**

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn is built on top of Matplotlib and is particularly well-suited for visualizing complex datasets with minimal code.

### Some key features of Seaborn include:

**High-level Functions**: Seaborn provides functions to create various types of statistical plots with just a few lines of code. This includes scatter plots, line plots, bar plots, histograms, box plots, violin plots, and more.

**Statistical Estimations**: Seaborn can also perform and display statistical estimation and aggregation within the plots, making it easier to understand the underlying data distribution.

**Integration with Pandas**: Seaborn works seamlessly with Pandas DataFrames, which is a popular data manipulation library in Python. You can easily create visualizations from DataFrame data.

**Customizable Aesthetics**: Seaborn comes with built-in themes and color palettes to improve the aesthetics of your plots. You can also easily customize the appearance of your plots.

**Facet Grids**: Seaborn supports creating multiple plots arranged in a grid, which can help you explore relationships in data more effectively.

**Regression Plots**: It includes functions for creating regression plots, which can help you visualize relationships between variables and fit regression models.

**Categorical Data**: Seaborn has specialized functions for working with categorical data, making it easy to create plots that show the distribution of categorical variables.

## Seaborn - Histogram

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.histplot(df['petal_length'],kde = True)#Kernel Density Estimates
plt.show()
```

**Kernel Density Estimates**

Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis.

## Visualizing Pairwise Relationship

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.set_style("ticks")
sb.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

## Multi Panel Categorical Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Filter the dataset to remove rows where the "deck" column is null
df_filtered = df[df['deck'].notnull()]

# Create a categorical plot with multiple subplots using catplot
sns.catplot(x="alive", col="deck", col_wrap=3, data=df_filtered, kind="count",palette="rocket")

# Show the plot
plt.show()
```

##Linear Relationships

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
df = sns.load_dataset('tips')

# Set a colorful palette
sns.set_palette("husl")

# Create a FacetGrid for more organized plots
g = sns.FacetGrid(df, col="time", hue="day", height=4)

# Create a regplot for each facet
g.map(sns.regplot, "total_bill", "tip", scatter_kws={"alpha":0.5})

# Add a legend
g.add_legend()

# Show the plot
plt.show()
```

## Pair Grid



```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
g = sb.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
g = sb.PairGrid(df)
g.map_upper(plt.scatter)
g.map_lower(sb.kdeplot, cmap = "Blues_d")
g.map_diag(sb.kdeplot, lw = 3, legend = False);
plt.show()
```

## Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
pivot_data = flights_data.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(pivot_data, annot=True, fmt="d", cmap="YlGnBu")

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap")

# Show the plot
plt.show()
```
